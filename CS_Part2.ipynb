### Clearly China is non speaking country. So USA --> GBR --> IND tops the list for checkpoint 3

```python
#as we need to convert/rename/reuse the category where the first instance is useful, so....
master_frame['primary_sector'] = master_frame.category_list.str.split('|',expand=True)[0]
master_frame
```

```python
#importing mapping data
mapping = pd.read_csv('mapping.csv',encoding='ISO-8859-1',sep=',')
#mapping
```

```python
mapping.isnull().sum()
```

```python
# removing the row with missing values, as part of cleaning
mapping = mapping[~pd.isnull(mapping['category_list'])]
mapping.isnull().sum()  #testing
#mapping    
```

```python
#we can see that in the original data provided somebody replaced 'na' with '0' which affected every field, including the category name, need to fix that first
# replacing '0' with 'na' handling enterprose 2.0 as well

mapping['category_list'] = mapping['category_list'].apply(lambda x: x.replace('0', 'na') if ('2.0' not in x) else x )

#mapping[mapping.category_list.str.contains('0')]             #testing 2.0 field

#mapping
```

###### Reshaping the mapping table(dataset) using melt()


https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.melt.html

Youtube link (really helpful :) )
https://www.youtube.com/watch?v=oY62o-tBHF4

```python
#melting it...,
mapping = pd.melt(mapping,id_vars=['category_list'],
                  var_name='main_sectors',
                  value_name='select')
```

```python
#selecting only the mapped ones, ignore others
mapping = mapping[(mapping['select']!=0)]
```

```python
mapping
```

```python
#finally merging it :)
mapped_finally = pd.merge(master_frame,mapping,left_on='primary_sector',right_on='category_list',how='inner')
```

```python
#dropping redundant and non necessary columns
mapped_finally = mapped_finally.drop(['category_list_y','select'],axis = 1) 
```

```python
#checkpoint 4 
mapped_finally
```

```python
#checkpoint 5:
```

```python
#this checkpoint must have D1,D2,D3 with top3 eng speaking countries with venture as their fundng type, so...,
#D1 for USA , ,

D1 = mapped_finally.loc[(mapped_finally['country_code'] == 'USA') & (mapped_finally['funding_round_type']=='venture')]
D1
```

#### Also below two columns added to showcase the below mentioned
###### [1.] The total number (or count) of investments for each main sector in a separate column
###### [2.] The total amount invested in each main sector in a separate column

Transform : Call func on self producing a DataFrame with transformed values and that has the same axis length as self.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.transform.html

```python
D1['total_investment_count']  = D1.groupby(['main_sectors']).main_sectors.transform('count')
D1['total_investment_amount'] = D1.groupby(['main_sectors']).raised_amount_usd.transform('sum')

#using variables after adding the same to dfs for more precise results

D1_investment_count  = D1.groupby(['main_sectors']).main_sectors.count()
D1_investment_amount = D1.groupby(['main_sectors']).raised_amount_usd.sum()
D1
```

```python
#D2 for Great Britain(GBR) ,,.. 

D2 = mapped_finally.loc[(mapped_finally['country_code'] == 'GBR') & (mapped_finally['funding_round_type']=='venture')]
D2['total_investment_count']  = D2.groupby(['main_sectors']).main_sectors.transform('count')
D2['total_investment_amount'] = D2.groupby(['main_sectors']).raised_amount_usd.transform('sum')

#using variables after adding the same to dfs for more precise results

D2_investment_count  = D2.groupby(['main_sectors']).main_sectors.count()
D2_investment_amount = D2.groupby(['main_sectors']).raised_amount_usd.sum()
D2
```

```python
#D3 for India(IND)) ,,.. 

D3 = mapped_finally.loc[(mapped_finally['country_code'] == 'IND') & (mapped_finally['funding_round_type']=='venture')]
D3['total_investment_count']  = D3.groupby(['main_sectors']).main_sectors.transform('count')
D3['total_investment_amount'] = D3.groupby(['main_sectors']).raised_amount_usd.transform('sum')

#using variables after adding the same to dfs for more precise results

D3_investment_count  = D3.groupby(['main_sectors']).main_sectors.count()
D3_investment_amount = D3.groupby(['main_sectors']).raised_amount_usd.sum()
D3
```

```python
# Checkpoint 5 Que 1 
print('Total number of Investments (count)(USA) :', D1_investment_count.sum())
print('Total number of Investments (count)(GBR) :', D2_investment_count.sum())
print('Total number of Investments (count)(INR) :', D3_investment_count.sum())
```

```python
# Checkpoint 5 Que 2 
print('Total amount of investment (USD)(USA) :', D1_investment_amount.sum())
print('Total amount of investment (USD)(GBR) :', D2_investment_amount.sum())
print('Total amount of investment (USD)(INR) :', D3_investment_amount.sum())
```
