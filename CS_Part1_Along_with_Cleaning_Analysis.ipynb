# 1. Cleaning Data

```python
import numpy as np
import pandas as pd
```

```python
companies = pd.read_csv('companies.txt', delimiter = "\t" , encoding="ISO-8859-1")
rounds2 = pd.read_csv("rounds2.csv" , encoding="ISO-8859-1")
```

```python
companies.head()
#companies.columns
```

```python
rounds2.head()
#rounds2.columns
```

```python
#looks like there are upper and lower case words, so uniqueness converting everything to lower() case

#companies['permalink'] = companies['permalink'].str.lower()
#companies['name'] = companies['name'].str.lower()
companies[['permalink','name']] = companies[['permalink','name']].apply(lambda x : x.astype(str).str.lower())
companies.head()
```

```python
#lower() case for company_permalink for rounds df as well 
#rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
rounds2[['company_permalink']] = rounds2[['company_permalink']].apply(lambda x : x.astype(str).str.lower())
rounds2.head()
```

```python
#rounds2.company_permalink.unique()
len(rounds2.company_permalink.unique())
```

```python
len(companies.permalink.unique())
```

```python
#encoding and decoding the stings with most suitable values for this  dataset
# At first only permalink was taken into consideration, but after looking at the geograpically diverse dataset the names also were affected so included name as well and processed as following

companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies['name'] = companies.name.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
len(companies.permalink.unique())
```

```python
#doing the same as above for rounds2 df as we need to encode and decode properly to get required fields
#rounds2['company_permalink'] = rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')

rounds2['company_permalink'] = rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
len(rounds2.company_permalink.unique())
```

# 2. After cleaning with respect to encoding and decoding we need to do further cleaning as following

```python
#checkpoint 1 
#Sl. No 1.
print(len(rounds2.company_permalink.unique()))

#Sl. No 2.
print(len(companies.permalink.unique()))

#Sl. No 3. --> could be the 'permalink' as ans

#Sl. No 4. present in rounds but not in companies   
print(len(rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']), :]))
```

```python
# merging the two dfs
# Sl. No 5.
master_frame = pd.merge(companies, rounds2, how="inner", left_on="permalink", right_on="company_permalink")
master_frame
```

```python
len(master_frame.index)
```

```python
# removing redundant/duplicate columns
# as both colmns used for merging are there(both unique though), hence removing one of them(as duplicate)
master_frame =  master_frame.drop(['company_permalink'], axis=1) 
```

```python
#percentage of missing column wise values

round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

Now its quite clear that the columns `funding_round_code` is not that of use since its with about 73%(72.9%) missing values, not desirable in analysis. Moreover from the business aim in place, the column like `homepage_url` can seem worthless to keep as well.

Hence dropping them as part of cleaning up.

```python
# dropping columns and cleaning up the master_frame df for good ;-)
master_frame = master_frame.drop(['funding_round_code', 'homepage_url'], axis=1)
master_frame.head()
```

```python
#checking again
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)

#now we are left out with following columns where the null values > 5%, need to have them cleaned as well, before going forward
```

```python
master_frame.raised_amount_usd.describe()
```

Please understand that the column `raised_amount_usd` is vital to this case as because we want to analyze.

Now will have to start with `raised_amount_usd`, since its too important, lets see if we can impute values or not , or will have to drop those rows having nulls from this column to comprehend.

For imputing we  need to consider parametes like mean, standerd deviation, median, min, max and much more,
The mean as shown above isaround USD 10M USD, but the std doesn't look good. The min and max can't even consider either, as they are poles apart :( . 

With this type of wide spread stats the only operation here could be dropping the NANs off, so now let's just remove the rows having NaNs in `raised_amount_usd`, as thought earlier.


```python
# removing the NANs
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame['raised_amount_usd'] = master_frame['raised_amount_usd']/1000000
master_frame.raised_amount_usd.describe()
#had to convert it to million of $ range
```
